```python
df1 = pd.merge(x_orders_df,x_sku_df, on = "SKU",how = "inner")
df1.rename(columns = {"ExternOrderNo":"Order ID"},inplace = True)
df1= df1.drop_duplicates()
df1.head()
```

```python
df1.shape
```

```python
df1["weight(kg)"] = df1["Weight (g)"]/1000
df1["Total Weight as per X(kg)"] = df1["Order Qty"] * df1["weight(kg)"]
df1.head(5)
```

```python
df2 = pd.DataFrame(df1.groupby("Order ID")["Total Weight as per X(kg)"].sum()).reset_index()
df2.head(5)
```

```python
def calculate_weight_slab(total_weight):
    whole_part, decimal_part = str(round(total_weight,1)).split('.')
    whole_part = int(whole_part)
    decimal_part = int(decimal_part)
    if decimal_part > 5:
        return whole_part + 1
    else:
        return whole_part + 0.5

# Apply the function to create the new column 'Weight_Slab'
df2['Weight_Slab as per X(kg)'] = pd.DataFrame(df2['Total Weight as per X(kg)'].apply(calculate_weight_slab))

# Display the DataFrame with the new column
df2.head(5)
```

```python
df = pd.merge(df2,c_invoice_df,on = "Order ID",how = "inner")
print(df2.shape,c_invoice_df.shape,df.shape)
df.head(5)
```

```python
df.rename(columns = {"Charged Weight":"Total Weight as per Courier Company (KG)"},inplace = True)
df.head(5)
```

```python
df['Weight_Slab as per Courier Company (KG)'] = pd.DataFrame(df['Total Weight as per Courier Company (KG)'].apply(calculate_weight_slab))

# Display the DataFrame with the new column
df.head(5)
```

```python
df.rename(columns = {"Zone":"Delivery Zone Charged By Courier Company"},inplace = True)
df.head(5)
```

```python
df.drop(columns=["Warehouse Pincode"],inplace = True)
df.head(5)
```

```python
df = pd.merge(df,x_pincode_df,on = "Customer Pincode",how = "inner")
print(x_pincode_df.shape,df.shape)
df.head(5)
```

```python
df= df.drop_duplicates()
df.shape
```

```python
df.rename(columns = {"Zone":"Delivery Zone as per X"},inplace = True)
df.head(5)
```

```python
df.rename(columns = {"Billing Amount (Rs.)":"Charges Billed By Courier Company (Rs.)"},inplace = True)
df.head(5)
```

```python
def cal_charges(row):
    slab_weight = row['Weight_Slab as per X(kg)']
    zone = row['Delivery Zone as per X']
    ship_type = row['Type of Shipment']
    fwd_fixed_rate = c_rate_df.loc[0,f'fwd_{zone}_fixed']
    fwd_add_rate = c_rate_df.loc[0,f'fwd_{zone}_additional']
    rto_fixed_rate = c_rate_df.loc[0,f'rto_{zone}_fixed']
    rto_add_rate = c_rate_df.loc[0,f'rto_{zone}_additional']
    
    if ship_type == 'Forward charges':
        total_charges = fwd_fixed_rate + fwd_add_rate*(slab_weight - 0.5)/0.5
            
    elif ship_type == 'Forward and RTO charges':
        fwd_charges = fwd_fixed_rate + fwd_add_rate*(slab_weight - 0.5)/0.5
        rto_charges = rto_fixed_rate + rto_add_rate*(slab_weight - 0.5)/0.5
        total_charges = fwd_charges + rto_charges
    else :
        total_charges = -99
    return total_charges
```

```python
df["Expected Charges as per X (Rs.)"] = df.apply(lambda x: cal_charges(x),axis = 1)
df.head(5)
```

```python
df["Difference Between Expected Charges and Billed Charges (Rs.)"] = df["Charges Billed By Courier Company (Rs.)"] - df["Expected Charges as per X (Rs.)"]
df.head(5)
```

```python
df.shape
```

```python
df.drop(columns=["Customer Pincode","Type of Shipment","Warehouse Pincode"],inplace = True)
df.head(5)
```

```python
arranged_columns = ["Order ID","AWB Code","Total Weight as per X(kg)","Weight_Slab as per X(kg)","Total Weight as per Courier Company (KG)",
                   "Weight_Slab as per Courier Company (KG)","Delivery Zone as per X","Delivery Zone Charged By Courier Company","Expected Charges as per X (Rs.)",
                   "Charges Billed By Courier Company (Rs.)","Difference Between Expected Charges and Billed Charges (Rs.)"]
df = df[arranged_columns]
```

```python
df.head(5)
```

```python
correctly_charged = df['Difference Between Expected Charges and Billed Charges (Rs.)'] == 0
overcharged = df['Difference Between Expected Charges and Billed Charges (Rs.)'] > 0
undercharged = df['Difference Between Expected Charges and Billed Charges (Rs.)'] < 0

charged_categories = ['Correctly Charged', 'Overcharged', 'Undercharged']
counts = [correctly_charged.sum(), overcharged.sum(), undercharged.sum()]
amounts = [
    df.loc[correctly_charged, 'Difference Between Expected Charges and Billed Charges (Rs.)'].sum(),
    df.loc[overcharged, 'Difference Between Expected Charges and Billed Charges (Rs.)'].sum(),  # Corrected to take the negative value
    df.loc[undercharged, 'Difference Between Expected Charges and Billed Charges (Rs.)'].sum()
]


result_df = pd.DataFrame({
    'Charged Category': charged_categories,
    'Count': counts,
    'Amount': amounts
})

print(result_df)
```

```python
with pd.ExcelWriter("Logistic.xlsx") as writer:
    df.to_excel(writer, sheet_name="Calculations")  
    result_df.to_excel(writer, sheet_name="Summary")
```
